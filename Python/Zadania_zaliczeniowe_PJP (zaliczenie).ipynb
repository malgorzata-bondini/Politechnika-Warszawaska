2. Stwórz słownik, gdzie kluczami są nazwy produktów (np. „jabłko”, „chleb”, „mleko”), a wartościami — ich ceny (liczby typu float). Następnie:

- Poproś użytkownika o nazwę produktu,
- Jeśli produkt istnieje, wypisz jego cenę,
- W przeciwnym razie poproś użytkownika o cenę, dodaj wpis do słownika i poinformuj o aktualizacji.



```python
import json
import os
import tkinter as tk
from tkinter import simpledialog, messagebox

root = tk.Tk()
root.withdraw()
```

```python
plik = "ceny.json"

if os.path.exists(plik):
    with open(plik, "r", encoding="utf-8") as f:
        ceny = json.load(f)
else:
    ceny = {
        "jabłko": 3.49,
        "chleb": 5.99,
        "mleko": 4.29
    }
```

```python
produkt = simpledialog.askstring("Produkt", "Podaj nazwę produktu:")
if produkt is None:
    messagebox.showinfo("Koniec", "Anulowano.")
else:
    produkt = produkt.strip().casefold()

    mapa_kluczy = {k.casefold(): k for k in ceny.keys()}

    if produkt in mapa_kluczy:
        klucz = mapa_kluczy[produkt]
        messagebox.showinfo("Cena", f"Cena: {ceny[klucz]:.2f} zł")
    else:
        tekst_ceny = simpledialog.askstring("Nowy produkt", "Nie mam tego produktu. Podaj cenę:")
        if tekst_ceny is None:
            messagebox.showinfo("Koniec", "Anulowano.")
        else:
            cena = float(tekst_ceny.strip().replace(",", "."))
            ceny[produkt] = cena

            with open(plik, "w", encoding="utf-8") as f:
                json.dump(ceny, f, ensure_ascii=False, indent=2)

            messagebox.showinfo("Zapisano", f"Dodano: {produkt} - {cena:.2f} zł")
            lines = [f"{name}: {price:.2f} zł" for name, price in sorted(ceny.items())]
            messagebox.showinfo("Słownik", "\n".join(lines))
```

3. Napisz funkcję obliczającą największy wspólny dzielnik (NWD) dla dwóch liczb całkowitych (algorytm Euklidesa). Program powinien pobrać dwie liczby od użytkownika, zweryfikować, że są dodatnie, a następnie wypisać wynik.


```python
def nwd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
a_text = input("Podaj pierwszą dodatnią liczbę całkowitą: ").strip()
b_text = input("Podaj drugą dodatnią liczbę całkowitą: ").strip()
```

```python
a = int(a_text)
b = int(b_text)
```

```python
if a <= 0 or b <= 0:
    print("Błąd: obie liczby muszą być dodatnie.")
else:
    print("NWD =", nwd(a, b))
```

4. Stwórz funkcję obliczającą iloczyn dla dowolnej liczby mnożników.

```python
def iloczyn(*liczby):
    wynik = 1
    for x in liczby:
        wynik *= x
    return wynik

print(iloczyn(2, 3))
print(iloczyn(2, 3, 4))
print(iloczyn(1.5, 2, 10.8))
```

Funkcja ma zwracać wartość: x^n + y^n.
Dodaj docstring opisujący działanie, typy argumentów i zwracaną wartość.

```python
def suma_poteg(y, x=2, n=2):
    """
    - Liczy sumę potęg dwóch potęg.
    - n to liczba całkowita, a x i z to liczny - int lub float.
    - Zwracany wynik to liczba typu int lub float w zależności od typów argumentów.
    """
    return x**n + y**n
```

10. Napisz program, który dla dowolnej liczby całkowitej dodatniej podanej przez użytkownika:

- Generuje ciąg Collatza (3x+1),
- Oblicza i wypisuje liczbę iteracji potrzebnych do osiągnięcia 1,
- Wypisuje maksymalną wartość, która pojawiła się w ciągu.

W problemie Collatza bierzemy dowolną liczbę $c_{0}$. Jeśli jest ona parzysta to $c_{1}$=$c_{0}$/2, a jeśli nieparzysta to $c_{1}$=3*$c_{0}$+1. Następnie kontynuujemy ten proces dla kolejnych liczb.  

Rekurencyjnie otrzymany w ten sposób ciąg liczb naturalnych określony jest przez następującą formułę:

$$
c_{n+1}=
\begin{cases}
\frac{1}{2}c_{n} & \text{gdy } c_{n} \text{ jest parzysta} \\
3c_{n}+1 & \text{gdy } c_{n} \text{ jest nieparzysta}
\end{cases}
$$


```python
import requests
from PIL import Image, ImageTk
from io import BytesIO

root = tk.Tk()
root.withdraw()
```

Obrazek do error handling dla użytkownika:

```python
url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQ5rFH6XhJwLkN_FhhwXJ-Em4X5FXdGnQ_K3g&s"

def popup_z_obrazkiem_z_url(tytul, wiadomosc, url):
    okno = tk.Toplevel(root)
    okno.title(tytul)
    okno.resizable(False, False)

    r = requests.get(url, timeout=10)
    img = Image.open(BytesIO(r.content))
    img = img.resize((240, 240))
    tk_img = ImageTk.PhotoImage(img, master=okno)

    tk.Label(okno, image=tk_img).pack(padx=12, pady=(12, 6))
    okno.image = tk_img

    tk.Label(okno, text=wiadomosc).pack(padx=12, pady=(0, 12))
    tk.Button(okno, text="OK", command=okno.destroy).pack(pady=(0, 12))

    okno.grab_set()
    okno.focus_set()
    okno.wait_window()
```

Error handling - dodałam dodatkowo

```python
while True:
    start_text = simpledialog.askstring("Collatz", "Podaj dodatnią liczbę całkowitą:")
    if start_text is None:
        messagebox.showinfo("Koniec", "Anulowano")
        start = None
        break

    start_text = start_text.strip()

    try:
        start = int(start_text)
    except ValueError:
        messagebox.showerror("Błąd", "Podaj liczbę całkowitą, spróbuj ponownie")
        popup_z_obrazkiem_z_url("Błąd", "To nie jest liczba całkowita", url)
        continue

    if start <= 0:
        messagebox.showerror("Błąd", "Podaj liczbę dodatnią, spróbuj ponownie")
        popup_z_obrazkiem_z_url("Błąd", "Liczba musi być dodatnia", url)
        continue

    break
```

```python
if start is not None:
    ciag = [start]
    n = start
    maks = start

    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        ciag.append(n)
        if n > maks:
            maks = n

    iteracje = len(ciag) - 1

    messagebox.showinfo("Wynik", f"Iteracje do 1: {iteracje}\nMaksymalna wartość: {maks}")
```

```python
if start is not None:
    if len(ciag) > 200:
        pokaz = ciag[:50] + ["..."] + ciag[-50:]
    else:
        pokaz = ciag

    messagebox.showinfo("Ciąg Collatza", "\n".join(str(x) for x in pokaz))
```
